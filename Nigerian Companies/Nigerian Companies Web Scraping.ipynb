```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#logging the url address

url = 'https://en.wikipedia.org/wiki/List_of_companies_of_Nigeria'

response =  requests.get(url)
```

```python
#initiating beautiful soup

soup = BeautifulSoup(response.text, 'html')
```

```python
#getting the table data

table = soup.find('table', class_ = 'wikitable sortable')
```

```python
#extracting the headers only

titles = table.find_all('th')
```

```python
#cleaning up the headers and removing unnnecessary characters

for row in titles:
    individual_title = [title.text.strip() for title in titles[:6]]
print(individual_title)
```

```python
#creating a dataframe with the columns using the header values

df = pd.DataFrame(columns = individual_title)
df
```

```python
#finding all table rows

data = table.find_all('tr')
```

```python
#extracting table details and adding to the dataframe

for row in data[2:]:
    row_data = row.find_all('td')
    individual_row_data = [datatitle.text.strip() for datatitle in row_data[:6]]
    length = len(df)
    df.loc[length] = individual_row_data
    print(individual_row_data)
```

```python
#previewing dataframe
df
```

```python
#exporting the dataframe into a csv

df.to_csv(r'Nigerian Companies.csv', index=False)
```
